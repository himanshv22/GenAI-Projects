**Building the Gemini Chatbot**

Step1:text streaming function

```python
def stream(text):
  print("Bot: ",end = ' ')
  words = text.split()#splits sentence into the list of words
  for word in words:
    print(word, end = ' ', flush =True)#forces output buffer to be flushed immediately meaning each word is shown on the screen right away rather thanm being delayed by the systems norms
    time.sleep(0.1)#after printing each word the function pauses breifly using this function, program waits for 0.1 seconds before continuing to the next word
  print()
```

```python
test = 'I am Building a Chatbot' # normal test case
stream(test)
```

**Step 2: connect to geminin API**

```python
try:
  client = genai.Client(api_key=key) #initialized the connection with Gemini API
  chat = client.chats.create(model = "gemini-2.5-flash") #sets up a chat session gemini's 2.5 flash model
except KeyError:
  print(f"Error with Gemini API key.\nError Message.\n{KeyError}") # to print message if API key is invalid or missing
```

**Step 3: create a funtion to talk to gemini**

```python
def ask_gemini(user_input): #function to take user input
  try:
    prompt = f"""You are Rose a  the poetical and lyrical AI that converses entirely in poems and poetic prose. Here's what you should do:

    1. You should have natural flow and rhythm and musical qualities
    2. Use alliterations, metaphors, similies and imagery through your senses
    3. Use personification for explaining abstract concepts
    4. Be in line with the poetic essence and never converse in plain prose.
    5. Write only 6 to 7 lines

    Here's the question: {user_input}"""
    response = chat.send_message(prompt) #sends question to gemini and gets answer
    return response.text
  except Exception as e:
    return f"Oops, something went wrong: {e}"
```

** Step 4: build a chatbot loop**

```python
print("рџЊё Hey , I am Rose your Poetic AI Chatbot!")
print("Type 'quit' to exit.\n")
while True:
    user_input = input("Me : ")
    if user_input.lower() == 'quit':
        print("Bot: Farewell, dear traveler of words. ")
        break
    response = ask_gemini(user_input)
    stream(response)
```
