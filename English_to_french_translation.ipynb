**Step 1: Install & Import**

```python
!pip install transformers sentencepiece
# transformers: Hugging Face library that gives us access to pre-trained NLP models
# sentencepiece: required by Helsinki-NLP translation model for tokenization (breaking text into pieces)
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM # importing pipeline, AutoTokenizer, AutoModelForSeqLM from transformers
import time #importing 'time' module to add delays between words for the typing effect
```

**Step 2: Load the Translation Model**

```python
# Helsinki-NLP is Hugging Face's best English→French model
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
# We are directly loading the model and tokenizer to bypass issues with the pipeline's task recognition.
```

**Step 3: Create stream() Function**

```python
def stream(text): #function to display translated text word by word simulating a natural typing effect
    print("Bot: ", end=' ')
    words = text.split()
    for word in words: #loops through each word in the translated sentence one by one
        print(word, end=' ', flush=True) #flush=True forces the output to appear immediately on screen
        time.sleep(0.1)
    print()
```

**Step 4: Translation Function:**

```python
def translate(user_input): #function that takes English text from the user and returns French translation
    try:
        # Tokenize the input text
        input_ids = tokenizer(user_input, return_tensors="pt").input_ids

        # Generate the translation
        outputs = model.generate(input_ids)

        # Decode the output IDs to get the translated text
        translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        return translated_text
    except Exception as e: # catches any unexpected errors
        return f"Oops, something went wrong: {e}"
```

**Step 5: Chatbot Loop**

```python
print("🌐 Welcome to the English-to-French Translation Chatbot!")
print("Type 'quit' to exit.\n")

while True:
    user_input = input("You (English): ")
    if user_input.lower() == 'quit': # checks if user typed 'quit'
        print("Bot: Au revoir! Goodbye!")
        break
    translation = translate(user_input) # calls the translate() function with the user's English text stores the returned French translation in the variable 'translation
    stream(translation) # calls the stream() function to display the French translation
```
